# Task-3 Demographic Analysis

#### Lets combine the dataframe by genre

```python
col = ['genre_1', 'genre_2','genre_3','CVotes10', 'CVotes09', 'CVotes08',
       'CVotes07', 'CVotes06', 'CVotes05', 'CVotes04', 'CVotes03', 'CVotes02',
       'CVotes01', 'CVotesMale', 'CVotesFemale', 'CVotesU18', 'CVotesU18M',
       'CVotesU18F', 'CVotes1829', 'CVotes1829M', 'CVotes1829F', 'CVotes3044',
       'CVotes3044M', 'CVotes3044F', 'CVotes45A', 'CVotes45AM', 'CVotes45AF',
       'CVotes1000', 'CVotesUS', 'CVotesnUS', 'VotesM', 'VotesF', 'VotesU18',
       'VotesU18M', 'VotesU18F', 'Votes1829', 'Votes1829M', 'Votes1829F',
       'Votes3044', 'Votes3044M', 'Votes3044F', 'Votes45A', 'Votes45AM',
       'Votes45AF', 'Votes1000', 'VotesUS', 'VotesnUS']
```

```python
df_genre = df.loc[:,col]
```

```python
df_genre
```

```python
df_genre.shape
```

```python
df_genre['cnt'] = 1
```

```python
df_by_g1 = df_genre.groupby('genre_1').sum()
df_by_g2 = df_genre.groupby('genre_2').sum()
df_by_g3 = df_genre.groupby('genre_3').sum()
```

```python
df_by_g1.head(2)
```

```python
df_by_g2.head(2)
```

```python
df_by_g3.head(2)
```

```python
df_add = df_by_g1.add(df_by_g2, fill_value=0).add(df_by_g3, fill_value=0)
```

```python
df_add
```

```python
genre_top10 = df_add[df_add['cnt']>10]
genre_top10
```

```python
genre_top10.iloc[:,:-1]=df_add.iloc[:,:-1].divide(df_add['cnt'], axis=0)
genre_top10
```

```python
genre_top10.loc[:,"VotesM":"VotesnUS"] = round(genre_top10.loc[:,"VotesM":"VotesnUS"],2)
```

```python
genre_top10.loc[:,"CVotes10":'CVotesnUS'] = round(genre_top10.loc[:,"CVotes10":'CVotesnUS'],0).astype(int)
```

```python
genre_top10.head(2)
```

```python
plt.figure(figsize=[20,20])
sns.barplot(df_add.index, df_add['cnt'])
plt.tick_params('x', rotation=45)
plt.xlabel('Genre')
plt.ylabel('Count of Movie')
plt.show()
```

#### Lets see difference in gender fb likes count

```python
male_grp = genre_top10.loc[:,['CVotesU18M','CVotes1829M','CVotes3044M','CVotes45AM']]
female_group = genre_top10.loc[:,['CVotesU18F','CVotes1829F','CVotes3044F','CVotes45AF']]
```

```python
fig, ax = plt.subplots(1,2, figsize=[20,10])
sns.heatmap(male_grp, annot=True, cmap='coolwarm', ax=ax[0])
sns.heatmap(female_group, annot=True,cmap='coolwarm', ax=ax[1])
plt.show()
```

```python
# People with age less than 18 and more than 45 have lowest count and not differed much by movie genre
# Even though drama genre have highest count of movies but still not counted by many
# Action, Sci-fi and thriller movies are liked by both but genre like romance, animation, drama are more popular in females 
# Males of age group 18-30 have most count and hence are active while females of only age group of around 18 are actively voting
```

#### Lets cross-analyze US and Non-Us Voters

```python
df.head()
```

```python
IFUS=[]
for i,j in df.iterrows():
    if j['Country']=='USA':
        IFUS.append('US')
    else:
        IFUS.append('non-US')
print(IFUS)
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
IFUS = pd.DataFrame(IFUS, columns=['IFUS'])
IFUS.head()
```

```python
df = pd.concat([df,IFUS], axis=1)
```

```python
df
```

```python
sns.catplot(x='IFUS', y='CVotesUS',data=df, kind='box', ax=ax[0])
plt.xlabel('US vs non-US movies')
plt.ylabel('US Voters')
sns.catplot(x='IFUS', y='CVotesnUS',data=df, kind='box', ax=ax[1])
plt.xlabel('US vs non-US movies')
plt.ylabel('Non-US Voters')
plt.show()
```

```python
# Non US voters have voted with almost same distribution for both country movies plus for non us movies they have wider distribution.
# US voters also share similar distribution but their median value is slightly higher for US movies. 
```

```python
sns.catplot(x='IFUS', y='VotesUS',data=df, kind='box', ax=ax[0])
plt.xlabel('US vs non-US movies')
plt.ylabel('US Voters')
sns.catplot(x='IFUS', y='VotesnUS',data=df, kind='box', ax=ax[1])
plt.xlabel('US vs non-US movies')
plt.ylabel('Non-US Voters')
plt.show()
```

```python
# Us voters have given better ratings for US movies while they follow normal distribution for non-US movies
# Non-US voters have given less rating to non-US movies as compared to US movies
```
